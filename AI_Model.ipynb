```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from joblib import dump
```

```python
file = pd.read_csv("precise_irrigation_dataset.csv")
```

```python
print(file.head())
```

```python
print(file.info())
```

```python
def plot(data,name):
    sns.histplot(data)
    plt.title(name)
    plt.ylabel("Frequency")
    plt.xlabel(name)
    plt.show()
```

```python
colum_names = ['Soil_Moisture', 'Temperature', 'Humidity', 'Rain_Forecast_2H', 'Rain_Forecast_8H', 
 'Rain_Forecast_16H', 'Rain_Forecast_1D', 'Rain_Volume', 'Wind_Speed', 'Solar_Radiation', 
 'Soil_Type', 'Soil_Perm', 'Soil_Depth', 'Crop_Type', 'Growth_Stage', 
 'Previous_Watering', 'Time_Since_Irrigation', 'Irrigation_Efficiency', 'Water_Deficit', 
 'Evapotranspiration', 'Cumulative_Rain_Impact', 'Time_of_Day', 'Season', 'Watering_Duration']
for i in range(24):
    plot(file[colum_names[i]],colum_names[i])
```

```python
x = file.drop(["Watering_Duration"],axis=1).values
y = file["Watering_Duration"].values
```

```python
from sklearn.model_selection import train_test_split as tts
x1,x2,y1,y2 = tts(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
x1 = norm.fit_transform(x1)
x2 = norm.transform(x2)
y1 = np.array(y1,dtype=float)
y2 = np.array(y2,dtype=float)
```

```python
y1.reshape(-1,1)
y2.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
RFR = RandomForestRegressor(n_jobs=-1,n_estimators=100,random_state=42)
```

```python
loss = np.zeros([100])
y1 = y1.flatten()
y2 = y2.flatten()
for epochs in range(100):
    RFR.fit(x1,y1)
    yp = RFR.predict(x1)
    loss[epochs] = mean_squared_error(y1,yp)
    print(f"loss is equal to {loss[epochs]}")
sns.lineplot(loss)
plt.title("Loss Plot")
plt.ylabel("Epochs")
plt.xlabel("Loss")
plt.show()
```

```python
dump(RFR,"SAI.joblib")
```
